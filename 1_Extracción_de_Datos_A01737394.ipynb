```python
# Importar librerias necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

#Importar librería para descargar archivo
from google.colab import files

CT = pd.read_csv("CountryTable.csv")
```

```python
CT
```

```python
#imprimir las primeras 5 filas
CT.head(5)
```

```python
#Filtro por objeto
Filtro1=CT[CT["continent"]=="Europe"]
Filtro1
```

```python
Filtro1.to_csv("Entregable_Filtro1.csv")
files.download("Entregable_Filtro1.csv")
```

```python
#Filtro por filas
Filtro2=CT.iloc[55:75, : ]
Filtro2
```

```python
Filtro2.to_csv("Entregable_Filtro2.csv")
files.download("Entregable_Filtro2.csv")
```

```python
#Filtro por columnas
Filtro3=CT.iloc[:, [2,4,6]]
Filtro3
```

```python
Filtro3.to_csv("Entregable_Filtro3.csv")
files.download("Entregable_Filtro3.csv")
```

```python
CT2=pd.read_csv("CountryTable.csv",index_col=2)
CT2
```

```python
#Filtro de filas contra columnas
Filtro4=CT2.loc[["Europe","Asia"],["code"]]
Filtro4
```

```python
Filtro4.to_csv("Entregable_Filtro4.csv")
files.download("Entregable_Filtro4.csv")
```

```python
#Filtro para mostrar las primeras 15 filas
Filtro5=CT2.head(15)
Filtro5
```

```python
Filtro5.to_csv("Entregable_Filtro5.csv")
files.download("Entregable_Filtro5.csv")
```

```python
#Filtro para solo mostrar la esperanza de vida mayor a 50
Filtro6=CT2[CT2["life_expectancy"]>50]
Filtro6
```

```python
Filtro6.to_csv("Entregable_Filtro6.csv")
files.download("Entregable_Filtro6.csv")
```

```python
#Filtro Y
Filtro7=CT2[(CT2["life_expectancy"] > 50) & (CT2["government_form"] == "Republic")]
Filtro7
```

```python
Filtro7.to_csv("Entregable_Filtro7.csv")
files.download("Entregable_Filtro7.csv")
```

```python
#Filtro O
Filtro8=CT2[(CT2["life_expectancy"] > 50) | (CT2["government_form"] == "Republic")]
Filtro8
```

```python
Filtro8.to_csv("Entregable_Filtro8.csv")
files.download("Entregable_Filtro8.csv")
```

```python
#Filtro not
Filtro9=CT2[(CT2["life_expectancy"] > 50) & ~(CT2["government_form"] == "Republic")]
Filtro9
```

```python
Filtro9.to_csv("Entregable_Filtro9.csv")
files.download("Entregable_Filtro9.csv")
```
